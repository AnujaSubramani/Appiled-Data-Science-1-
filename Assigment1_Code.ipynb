```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_dataset(file_path):
    """
    Load the dataset from a CSV file.

    Parameters:
    - file_path (str): The file path of the CSV file.

    Returns:
    - DataFrame: The loaded dataset.
    """
    return pd.read_csv(file_path)

def remove_columns(dataframe, columns):
    """
    Remove specified columns from the DataFrame.

    Parameters:
    - dataframe (DataFrame): The DataFrame from which columns will be removed.
    - columns (list): List of column names to be removed.
    """
    dataframe.drop(columns=columns, axis=1, inplace=True)
    print(dataframe)

def plot_price_histogram(dataframe):
    """
    Create a histogram of car prices.

    Parameters:
    - dataframe (DataFrame): The DataFrame containing car data.
    """
    plt.hist(dataframe['price'], color='blue', edgecolor='black', bins=10)
    plt.title('Range of Car Prices in USA')
    plt.xlabel('Prices')
    plt.ylabel('Number of Cars')
    plt.show()

def plot_price_vs_mileage(dataframe):
    """
    Create a scatter plot of car prices vs. mileage.

    Parameters:
    - dataframe (DataFrame): The DataFrame containing car data.
    """
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x=dataframe['price'], y=dataframe['mileage'])
    plt.title('Prices vs Mileage of Cars in USA')
    plt.xlabel('Prices')
    plt.ylabel('Mileage')
    plt.show()

def plot_price_by_brand(dataframe):
    """
    Create a box plot of price range across car brands.

    Parameters:
    - dataframe (DataFrame): The DataFrame containing car data.
    """
    plt.figure(figsize=(12, 6))
    sorted_brands = dataframe.groupby(
        ['brand'])['price'].median().sort_values()
    sns.boxplot(x=dataframe['brand'], y=dataframe['price'],
                order=list(sorted_brands.index))
    plt.title('Price Range Across Car Brands in USA')
    plt.xlabel('Brand')
    plt.ylabel('Price')
    plt.xticks(rotation=70)
    plt.show()

def display_kurtosis(dataframe):
    """
    Display the kurtosis for numeric columns in the DataFrame.

    Parameters:
    - dataframe (DataFrame): The DataFrame containing numeric columns.
    """
    numeric_columns = dataframe.select_dtypes(include=['number']) # Select numeric columns
    print(numeric_columns.kurtosis())

def main():
    """
    Main function to load the dataset, process it, and plot visualizations.
    """
    file_path = r"/content/USA_cars_datasets.csv"
    df = load_dataset(file_path)
    columns_to_remove = ['Unnamed: 0', 'vin']
    remove_columns(df, columns_to_remove)

    plot_price_histogram(df)
    plot_price_vs_mileage(df)
    plot_price_by_brand(df)
    display_kurtosis(df) # Add display of kurtosis

main()
```
